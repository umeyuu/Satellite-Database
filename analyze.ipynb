```python
import pandas as pd
```

```python
df = pd.read_csv('charge.csv', parse_dates=['date'])
```

```python
def sum_ten(df):
    columns = ['one_minute_after', 'two_minute_after', 'three_minute_after','four_minute_after', 'five_minute_after', 
               'six_minute_after', 'seven_minute_after', 'eight_minute_after', 'nine_minute_after', 'ten_minute_after']
    ans = 0
    for col in columns:
        ans += df[col]
    return ans

df['ten_label'] = df.apply(sum_ten, axis=1)
df.set_index('date', inplace=True)
```

```python
df.ten_label.plot.hist(bins=50)
```

```python
df[df.ten_label >= 10].ten_label.plot.hist(bins=50)
```
